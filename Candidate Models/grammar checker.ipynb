```python
import pandas as pd
df = pd.read_csv('./extra_data.csv')
```

```python
def check_grammar(txt):
    text =txt
    matches = tool.check(text)
    return matches, len (matches)
```

```python
feed_list = []
for i in df.iloc[:,1]:
    txt, cnt = check_grammar(i)
    feed_list.append ([i, txt, cnt])
```

```python
data = pd. DataFrame(feed_list)
```
